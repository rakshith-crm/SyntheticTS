```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
print(tf.__version__)
```

```python
def plot_series(time, series, format='-',title=None, start=0, end=None):
    plt.figure(figsize=(7, 4))
    plt.plot(time[start:end], series[start:end], format)
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.grid(True)
def MAE(y_valid, forecast):
    return keras.metrics.mean_absolute_error(y_valid, forecast)
def MSE(y_valid, forecast):
    return keras.metrics.mean_squared_error(y_valid, forecast)

def trend(time, slope=0):
    return slope * time

def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)

def noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
# Creating the time series
baseline = 10
amplitude = 40
slope = 0.05
noise_level = 5

time = np.arange(4*365 +1, dtype='float32')
# Adding trend and seasonality
series = baseline + trend(time, slope) + seasonality(time, period=period, amplitude=amplitude)

# Adding noise
series += noise(time, noise_level=noise_level)
# plotting the time series
plot_series(time, series)
```

```python
# Splitting into train and valid
split_time = 1000
X_train = time[:split_time]
y_train = series[:split_time]
X_valid = time[split_time:]
y_valid = series[split_time:]
plot_series(X_train, y_train, title='Train')
plot_series(X_valid, y_valid, title='Valid')
```

## Utility function to print predictions, validation and statistics such as mean abs error and mean square error


```python
def model_performance(X_valid, y_valid, prediction, model_name=None):
    plot_series(X_valid, y_valid)
    plt.plot(X_valid, prediction)
    plt.title(model_name)
    plt.figure(figsize=(18, 9))
    print('MAE : {}, MSE : {}'.format(MAE(y_valid, prediction),MSE(y_valid, prediction)))
```

# Naive Forecasting
Perdicting the next value to be the same as previous value

```python
naive_forecast = series[split_time-1:-1]
model_performance(X_valid, y_valid,naive_forecast, model_name='Naive Forecast')
```

```python
plt.figure(figsize=(10, 6))
plot_series(X_valid, y_valid, start=0, end=150)
plt.plot(X_valid[1:151], naive_forecast[1:151])
```

```python
print('Naive forecast:\nMAE : {}, MSE : {}'.format(MAE(y_valid, naive_forecast),MSE(y_valid, naive_forecast)))
```

```python
def moving_average(series, window_size=10):
    forecast = []
    for time in range(len(series)-window_size):
        forecast.append(series[time : time+window_size].mean())
    return np.array(forecast)
```

```python
window_size = 30
moving_avg = moving_average(series, window_size=window_size)

model_performance(X_valid, y_valid, moving_avg[split_time-window_size:])
```

## Removing Trend and seasonality

```python
diff_series = (series[365:] - series[:-365])
diff_time = time[365:]

# Now we have removed trend and seasonality 
plot_series(diff_time, diff_series)
```

Now that we have eliminated trend and seasonality, we can do moving average on this data

```python
diff_moving_avg = moving_average(diff_series, 50)
diff_moving_avg = diff_moving_avg[split_time-365-50:]

plot_series(X_valid, diff_series[split_time-365:])
plt.plot(X_valid, diff_moving_avg)
```

```python
diff_moving_avg_plus_past = series[split_time - 365 : -365] + diff_moving_avg

model_performance(X_valid, y_valid, diff_moving_avg_plus_past)
```

```python
diff_moving_avg_plus_smooth_past = moving_average(series, 10)[split_time - 370:-360] + diff_moving_avg

model_performance(X_valid, y_valid, diff_moving_avg_plus_smooth_past)
```
